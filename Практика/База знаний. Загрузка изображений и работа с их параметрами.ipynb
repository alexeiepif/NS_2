## Подгрузка библиотек и тестового изображения

```python
# Для загрузки файла в Colab
import gdown

# Для отрисовки изображений
from PIL import Image

# Для отрисовки графиков
import matplotlib.pyplot as plt

# Импорт библиотеки NumPy
import numpy as np
```

# Загружаем тестовое изображение

```python
gdown.download('https://ia601509.us.archive.org/view_archive.php?archive=/2/items/CAT_DATASET/CAT_DATASET_01.zip&file=CAT_00%2F00000001_000.jpg', 'cat.jpg', quiet = True)
```

# загружаем изображение в ноут и смотрим размеры изображения

```python
image = Image.open('cat.jpg')        # Открываем изображение
width = image.size[0]                # Определяем ширину
height = image.size[1]               # Определяем высоту

print('параметры изображения:')
print('ширина изображения', width)
print('высота изображения', height)
print('формат изображения', image.format)
```

**Смотрим само изображение**

```python
plt.imshow(image)
plt.show
```

# Можно изменить размер изображения

```python
new_size = (28,28)
image.thumbnail(new_size, reducing_gap=None)
saved = "resize_cat.jpeg"
image.save(saved)
plt.imshow(image)
plt.show

width = image.size[0]                 # Определяем ширину
height = image.size[1]                # Определяем высоту

print('параметры изображения:')
print('ширина изображения', width)
print('высота изображения', height)
print('формат изображения', image.format)
```

# Перевод в Grey

```python
image = Image.open('cat.jpg').convert('LA')
image.save('grey_cat.png')

image = Image.open('grey_cat.png')             # Открываем изображение
width = image.size[0]                          # Определяем ширину
height = image.size[1]                         # Определяем высоту

plt.imshow(image)
plt.show

print('параметры изображения:')
print('ширина изображения', width)
print('высота изображения', height)
print('формат изображения', image.format)
```

# Из изображения в массив

```python
image = Image.open('cat.jpg')
im2arr = np.array(image)                # массив с размерами: height x width x channel
print('форма массива', im2arr.shape)
```

# Из массива в изображение

```python
arr2im = Image.fromarray(im2arr)

width = arr2im.size[0]              # Определяем ширину
height = arr2im.size[1]             # Определяем высоту

plt.imshow(arr2im)
plt.show

print('параметры изображения:')
print('ширина изображения', width)
print('высота изображения', height)
```
